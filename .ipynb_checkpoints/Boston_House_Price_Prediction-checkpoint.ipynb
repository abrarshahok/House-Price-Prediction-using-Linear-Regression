# Boston Housing Price Prediction - Linear Regression

## Boston Housing Dataset

The Boston Housing dataset is used for predicting housing prices. It includes the following features:

- **CRIM**: Crime rate by town
- **ZN**: Proportion of land zoned for large lots
- **INDUS**: Proportion of non-retail business acres
- **CHAS**: Charles River (1 if tract bounds river, 0 otherwise)
- **NOX**: Nitric oxides concentration
- **RM**: Average number of rooms per dwelling
- **AGE**: Proportion of units built before 1940
- **DIS**: Distance to Boston employment centers
- **RAD**: Accessibility to radial highways
- **TAX**: Property tax rate
- **PTRATIO**: Pupil-teacher ratio
- **B**: \(1000(Bk - 0.63)^2\) where Bk is the proportion of Black residents
- **LSTAT**: Percentage of lower status population
- **MEDV**: Median value of homes ($1000s)

```python
# Importing neccessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

## Making Data ready for Training

### Accessing Data Set

```python
# Path to the dataset
path = './BostonHousing.csv'
# Reading the dataset into a pandas DataFrame
data = pd.read_csv(path)
```

## Ploting Correlation Heatmap

```python
correlation = data.corr()

cmap = LinearSegmentedColormap.from_list("custom_palette", ["green", "yellow", "orange"])

plt.figure(figsize=(10, 10))
sns.heatmap(correlation, 
            cbar=True, 
            square=True, 
            fmt='.1f', 
            annot=True, 
            annot_kws={'size': 8}, 
            cmap=cmap) 

plt.title('Correlation Heatmap', size=16)
plt.show()
```

### Splitting Dataset into X(Input Features), y(Output Labels)

```python
# Selecting all columns except the last one as features (X)
X = data.iloc[:, :-1].values
# Selecting the last column as the target variable (y)
y = data.iloc[:, -1].values
```

Normalizing data using Z-Score Normalization

```python
# Initializing the StandardScaler for normalization
scaler = StandardScaler()
# Fitting the scaler on the feature data and transforming it
X_norm = scaler.fit_transform(X)

# Splitting the normalized data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.18, random_state=46)
```

## Training Model

```python
# Initializing the SGDRegressor with a maximum of 10,000 iterations and tolerance of 1e-3
sgdr = SGDRegressor(max_iter=10000, tol=0.0001, alpha=0.01, learning_rate='adaptive')
# Training the SGDRegressor model on the training data
sgdr.fit(X_train, y_train)
```

```python
# Predicting the target values for the test set
y_pred = sgdr.predict(X_test)

# Printing the first 5 actual values from the test set
print(f"Actual values:\n{y_test[:5]}")
# Printing the first 5 predicted values
print(f"Predicted values:\n{y_pred[:5]}")
```

```python
# Evaluating the model performance
score = r2_score(y_test, y_pred)
print(f"R² score: {score}")
print(f"The accuracy of our model is {round(score * 100, 2)}%")
```

## Plotting target vs predicted values

```python
# Number of features
num_features = X.shape[1]

# Create a figure with a grid of subplots
# Calculate number of rows and columns for the grid
ncols = 3  # Number of columns
nrows = (num_features + ncols - 1) // ncols  # Number of rows

# Create a figure with the calculated grid
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 5 * nrows))

# Flatten the axes array for easy iteration
axes = axes.flatten()

for i in range(num_features):
    axes[i].scatter(X_test[:, i], y_test, alpha=0.5, color='orange', label='Target', s=50)
    axes[i].scatter(X_test[:, i], y_pred, alpha=0.5, color='green', label='Predicted',s=50)
    axes[i].set_xlabel(data.columns[i].title())
    axes[i].set_ylabel('Value')
    axes[i].set_title(f'{data.columns[i].title()} (Target vs Predicted)')
    axes[i].legend()

# Hide any unused subplots
for j in range(num_features, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

print(len(y_test), len(y_pred))

plt.scatter(range(len(y_test)), y_test, color='green', label='Targeted Price', alpha=0.6, s=100)

plt.scatter(range(len(y_pred)), y_pred, color='orange', label='Predicted Price', alpha=0.6, s=100)

plt.title('Target vs. Predicted Price')
plt.xlabel('Index')  
plt.ylabel('Price')  
plt.legend()
plt.grid()
plt.show()
```
